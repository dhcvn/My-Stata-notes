<h4>Question: Analysis of drug consumption is with related factors?</h4>

```stata
// This dataset is the panel data
// 11 years from 1985 to 1995
// 48 states reported each year

// Declare panel data in Stata and summary data
xtset id year
xtdes
xtsum $ylist $xlist
```

```stata
// Differents model to estimate relation between drug consumption with associate factors
qui eststo model1: reg $ylist $xlist //Pooled regression model

qui eststo model2: xtreg $ylist $xlist, be //Between effect
qui eststo model3: xtreg $ylist $xlist, fe //Fixed effect
qui eststo model4: xtreg $ylist $xlist, re //Random effect

esttab model1 model2 model3 model4, ci r2 noconstant //Showing model's result
reg D.($ylist $xlist), noconstant //First differents
hausman model3 model4
```
